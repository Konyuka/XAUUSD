# Baseline Performance Insights

Visualizes equity curve and key stats generated by `goldbot.reporting.baseline_report`.


```python
import json
from pathlib import Path

import pandas as pd

stats = json.loads(Path("reports/baseline_stats.json").read_text())
quality = json.loads(Path("reports/baseline_quality.json").read_text())
quality, stats
```

```python
from goldbot.data import load_cached_prices
from goldbot.features import engineer_feature_set
from goldbot.pipeline import run_baseline_backtest

prices = load_cached_prices()
features = engineer_feature_set(prices)
outcome = run_baseline_backtest()
outcome.backtest.equity_curve.tail()
```

```python
import plotly.express as px

fig = px.line(
    x=outcome.backtest.equity_curve.index,
    y=outcome.backtest.equity_curve.values,
    labels={"x": "Date", "y": "Equity (USD)"},
    title="Baseline Equity Curve",
)
fig
```
